#Identificación de Mejoras a través de la Información de Reclamaciones y Quejas

Dentro de la prestación del servicio de Energía Eléctrica en el país, los comercializadores de energía son los encargados de comprar y vender energía eléctrica tanto en el mercado mayorista como también a los usuarios finales; por tal motivo, en el proceso de esta transacción se debe garantizar una calidad optima de negocio que le garantice la mayor satisacción al cliente.

Por lo anterior, los comercializadores tienen como obligación recibir todas las reclamaciones y quejas que tienen los usuarios relacionadas con la prestación del servicio. 

Con este proyecto se busca identificar dentro de la información reportada por los prestadores, cuales son los procesos a mejorar dentro de la empresa, de acuerdo con la tipificación que se almacena en la información.

##¿Qué espera ganar su empresa con este proyecto?

De acuerdo con la información que recolecta la empresa relacionada a las reclamaciones y quejas de los usuarios, se podrán identificar las áreas en la cuales se tendrán opciones de mejora en los procesos. Considerando que, la tipificación con la que se recibe la información se puede identificar si las falencias en los procesos se encuentran en la facturación, la prestación del servicio, servicio al cliente, entre otras cosas.

así bien, al identificar los procesos donde se presenten el mayor número de reclamaciones y quejas, se podrán focalizar procesos de la empresa para mejorar la prestación del servicio y la percepción por parte de los usuarios

##¿Cómo definiría usted la culminación exitosa de nuestros esfuerzos?

Si al finalizar de analizar la información con la que cuenta la empresa, se pueden identificar los procesos en los cuales se focalizan la mayor parte de reclamaciones y quejas recibidas, y si con base en esto, se pueden ejecutar proyectos de mejora, se podrá dar como éxitoso el proceso inicial del proyecto; sin embargo, la culminación del proyecto se finalizará con éxito cuando se visualice una disminución significativa de las Reclamaciones y Quejas recibidas en la compañía.

Si se tiene el presupuesto para el ejecutar el proyecto. Inicialmente, solo se ha identificado costos a nivel de infrastructura y hora hombre, lo cual a la fecha se tiene en la empresa. Por tal motivo, se podría identificar la necesidad para incluir dentro de las actividades de los trabajadores.
