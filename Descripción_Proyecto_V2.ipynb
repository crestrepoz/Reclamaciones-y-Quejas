#Identificación de Mejora en los procesos a través de la Información de Reclamaciones y Quejas

##**Introducción**

El actual trabajo tiene como objetivo, identificar inicialmente dentro de una organización, como es el comportamiento de las solicitudes de Reclamaciones y Quejas que ingresan a la compañia, identificando de acuerdo con la regulación actual, cuales son los procesos que se ven mas con miran a las falencias de la prestación del servicio de energía eléctrica a los usuarios.

Durante la prestación del servicio de energía eléctrica, encontramos la comercialización como la actividad de comprar y vender energía, ya sea en el mercado mayorista o en el mercado minorista, este último segmento del mercado, se encuentra asociado directamente a la atención de los usuarios residenciales en el país; los cuales, de acuerdo con la regulación actual, tienen el derecho de presentar todas las inconformidades que tengan en cuanto a facturación, prestación e instalación del servicio.

Por lo anterior, las empresas comercializadoras tienen como obligación disponer de los medios adecuados para que los usuarios puedan interponer las solictudes. A su vez, está información puede ser utilziada por los entes de control, para identificar cual es la percepción que tienen los usuarios en cuanto a la prestación del servicio, pero a su vez, también es una manera de realizar actividades de inspección, vigilancia y control, buscando garantizar que al usuario se le entregue un servicio de calidad. 


De igual manera, para cada compañía, la información de reclamaciones y quejas, debería servir como un indicador individual para las áreas internas de la empresa; considerando que, la clasificación e identificación de las solicitudes, pueden ayudar a, inicialmente plantear y organizar opciones de mejora en las áreas asociadas a la mayoría de la solicitudes que ingresan, segundo a realizar campañas con miras a una la implementación de una mejor imagen ante los usuarios atendidos y así tener un mayor control de los esfuerzos que se deben realizar a nivel organizacional.

##**Planteamiento inicial del problema**

De toda la información que procesa la empresa en cuanto a reclamaciones, no se realiza una clasificación de la información, donde se pueda identificar las áreas en las cuales se deben enfocar los esfuerzos para brindar una mayor cálidad en cuento a la prestación del servicio de energía eléctrica. Entendiendo la prestación como todo el proceso desde la prestación del servicio y la facturación del mismo.

Por lo anterior, es posible que la empresa no esté teniendo en cuenta la percepción que tienen los usuarios en cuanto a la caldiad de los procesos,  de acuerdo con de acuerdo con el número de reclamaciones y quejas que se reciben, y a su vez tampoco se está realizando un enfoque especifico sobre en que áreas se deben fortalecer los esfuerzos para reducir la cantidad de reclamaciones y quejas que ingresan a la empresa.

Entendiendo así que entre mayor sea el número de solicitudes que ingresen los usuarios, se entienden como una opción de mejora que debe tener en cuenta la empresa. Un claro ejemplo de como se pueden asociar estas reclamaciones a los proceso es de acuerdo a las causas por las cuales, los usuarios interponen una solicitud. Actualmentes están las causales asociadas a facturación (Solicitudes tipo F) y solicitudes asociadas a la prestación del servicio (Solicitudes tipo P).


**Solicitudes de Facturación:** Estas solicitudes están asociadas en su mayoria a los procesos que se realizan en la Dirección Comercial de la empresa (Gestión de nuevos clientes, Lectura, facturación y reparto, Gestión de cobro y recaudo, Atención al cliente, Compras de energía, Gestión órdenes de servicio).

**Solicitudes de Prestación:** Estas solicitudes están asociadas a los procesos misionales de la empresa (Planificación de red, Desarrollo, Operación, Mantenimiento, Gestión de datos de red, Control de energía, Calidad del servicio).

En el anexo de la [Resolución SSPD No 20188000076635 del 2018](http://www.sui.gov.co/web/content/download/3075/26149/version/1/file/20188000076635_Final_Firmada_DIGITALIZADA%2B%281%29.pdf), se encuentra el detalle y nombre de la causal asociada la tipificación indicada; por tal razón, de acuerdo con está clasificación, se realizará un enfoque a las áreas que deben intervenir en el proceso.

##¿Qué espera ganar su empresa con este proyecto?

De acuerdo con la información que recolecta la empresa relacionada a las reclamaciones y quejas de los usuarios, se podrán identificar las áreas en la cuales se tendrán opciones de mejora en los procesos. Considerando que, la tipificación con la que se recibe la información se puede identificar si las falencias en los procesos se encuentran en la facturación, la prestación del servicio, servicio al cliente, entre otras cosas.

así bien, al identificar los procesos donde se presenten el mayor número de reclamaciones y quejas, se podrán focalizar procesos de la empresa para mejorar la prestación del servicio y la percepción por parte de los usuarios.

Actualmente, durante todas las actividades de prestación del servicio de energía electrica, participan al rededor de 15 áreas asociadas en la prestación y facturación del servicio, buscando garantizar inicialmente que a los usuarios se les facture bajo la premisa del consumo realizado por cada usuatio, pero a su vez que se facture por un servicio de buena calidad; por tal razón, al identificar cuales son las áreas en las cuales se asocian la mayor parte de las solicitudes que interponen los usuarios, garantizará que la empresa, como se mencionó anteriormente, realice una correcta asignación de recursos para la recuducción de la solicitudes que ingresan los ususarios.

En linea con lo anterior, se puede hablar de recursos asociados a las áreas que se encuentran asociadas a las solicitudes, tanto a nivel econónomico, como también en el tiempo que van a implementar en la solución de la solicitudes. Una de las prinicipales dificultades a nivel corporativo, es la cantidad de tiempo que debe asignar la empresa para solucionar todos los incovenientes con miras al usuario.

Así bien, la empresa estaría optimizando la organización a nivel de respuesta de las solicitudes entrantes, pero también la asiganción de recursos.



##¿Cómo definiría usted la culminación exitosa de nuestros esfuerzos?

Si al finalizar de analizar la información con la que cuenta la empresa, se pueden identificar los procesos en los cuales se focalizan la mayor parte de reclamaciones y quejas recibidas, y si con base en esto, se pueden ejecutar proyectos de mejora, se podrá dar como éxitoso el proceso inicial del proyecto; sin embargo, la culminación del proyecto se finalizará con éxito cuando se visualice una disminución significativa de las Reclamaciones y Quejas recibidas en la compañía.

Adicionalmente, si este proceso logra disminuir el tiempo de respuesta a las solicitudes que ingresan a las distintas áreas, sería un éxito en cuanto a la percepción que tienen los usuario ante la empresa, dado a que el tiempo de respuesta que perciben los usuarios está por encima de los tiempos dados por regulación y esto ha hecho que la empresa tenga una presencia negativa en el mercado.

Aunado a lo anterior, este proceso ayudaría a la empresa a dar cumplimiento a los tiempos dados por la leyes actuales en cuanto a la respuesta de los derechos de petición interpuestos por los usuarios.

### Descripción de los datos a utilizar

para este ejercicio partícular, se utilizará una base con corte a octubre del año 2020, estructurada en 16 columnas y 66.199 registros. 

De la base de datos utilizadas, hay registros que son de obligatorio registro, de acuerdo con lo establecido en la norma que exige el cargue; sin embargo, encontramos variables que no se obligatorio reporte, por lo cual encontraremos las siguientres variables con datos incompletos:

- Fecha de Respuesta (31204 Registros faltantes)
- Radicado de Respuesta (31204 Registros faltantes)
- Fecha de Notificación (59351 Registros faltantes)
- Fecha de Traslado a la SSPD (64956 Registros faltantes)

Por lo anterior, estas variables fueron excluidos del proceso final del análisis de datos, dado a que no eran significativos para el proceso de clasificación de solicitudes.

###Principales Resultados

- De la información analizada, se encontró que en el departamaneto del Atlántico es donde mas se presentan solicitudes de reclamaciones, quejas y recursos de reposición con un total de 46.109, seguido del departamento de Magdalea y La Guajira con 13.711 y 6374 respectivamente.
